```python
## Next, we'll import pyplot
import matplotlib.pyplot as plt
```

```python
## Creating a line Plot
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y Axis')
plt.title("Basic Line Plot")
plt.show()
```

```python
## Creating a customised line Plot
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3,markersize=9)
plt.grid(True)

```

```python
## Another Customized plot
a = [1, 2, 3, 4, 5]
b = [10, 12, 8, 14, 17]

plt.plot(
    x, y,
    color='purple',
    linestyle='-',
    linewidth=3,
    marker='D',
    markersize=10,
    markerfacecolor='orange',
    markeredgecolor='black',
    label='Sales Line'
)

plt.title('Company Sales Trend', fontsize=18, color='darkblue')
plt.xlabel('Quarter', fontsize=14)
plt.ylabel('Revenue ($k)', fontsize=14)

plt.grid(True, linestyle='--', color='grey', alpha=0.6)
plt.legend(loc='lower right')

plt.show()

```

```python
## Creating Multiple Plots
## Our Sample data
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

plt.figure(figsize=(9,5))   #This specifies the size of the plot wrapper where plots are contained

plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(y1,x,color='red')
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x,y2,color='blue')
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(x,y2,color='green')
plt.title("Plot 4")
```

```python
### Creating a Bar Plot
categories=['A','B','C','D','E']
values=[5,7,3,8,6]

plt.bar(categories,values,color='darkblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```
